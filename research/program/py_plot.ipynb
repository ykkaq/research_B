```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Data points
x = np.array([50, 100, 150, 200])
y = np.array([0.22815114629236252, 0.11455533660051737, 0.07655718822651922, 0.05749210273025131])

# Define exponential function for fitting
# Function of the form: y = a * exp(b * x)
def exp_func(x, a, b):
    return a * np.exp(b * x)

# Perform curve fitting to find optimal parameters
popt, pcov = curve_fit(exp_func, x, y)
a, b = popt

# Generate points for smooth curve plotting
x_smooth = np.linspace(min(x), max(x), 100)
y_smooth = exp_func(x_smooth, a, b)

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', label='Data points')
plt.plot(x_smooth, y_smooth, color='blue', label='Exponential fit')

# Customize plot appearance
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Points and Exponential Fit')
plt.legend()
plt.grid(True)

# Calculate and display the fitted equation parameters
print(f'Fitted equation: y = {a:.6f} * exp({b:.6f}x)')

# Calculate R-squared (coefficient of determination)
y_pred = exp_func(x, a, b)
residuals = y - y_pred
ss_res = np.sum(residuals ** 2)
ss_tot = np.sum((y - np.mean(y)) ** 2)
r_squared = 1 - (ss_res / ss_tot)
print(f'R-squared (R²): {r_squared:.6f}')

plt.show()
```
